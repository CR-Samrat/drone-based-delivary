```python
coordinates = (40.7128, -74.0060)  # Example: New York City

# Retrieve the street network for the specified location within a radius of 500 meters
G = ox.graph_from_point(coordinates, network_type='all')

# Plot the street network
ox.plot_graph(ox.project_graph(G))
plt.show()
```

```python
background = mpimg.imread('./maps/map.png')

# Define the extent of the image
extent = [22.5898728, 22.5917280, 88.4125104, 88.4131129]

# Calculate the aspect ratio of the extent
extent_width = extent[3] - extent[2]
extent_height = extent[1] - extent[0]
aspect_ratio = extent_width / extent_height

# Plot the image with adjusted aspect ratio
plt.imshow(background, extent=extent)
plt.gca().set_aspect(aspect_ratio, adjustable='box')
plt.show()
```
