```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./last_mile_performance_improvement.csv")

df.head()
```

```python
df_gec = df[df['manhattan_points'] == 0]
df_gmm = df[df['euclidean_points'] == 0]
df_ecmb = df[(df['euclidean_points'] > df['manhattan_points']) & (df['manhattan_points'] != 0)]
df_mmeb = df[(df['euclidean_points'] <= df['manhattan_points']) & (df['euclidean_points'] != 0)]
```

## GEC

```python
df_gec_50 = df_gec[df_gec['total_delivary'] == 50]
df_gec_50 = df_gec_50.sort_values(by="distance").reset_index()
df_gec_50.head(10)
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(data=df_gec_50, x=df_gec_50.index, y='distance', marker='o', label='GEC Approach')
sns.lineplot(data=df_gec_50, x=df_gec_50.index, y='optimized_distance', marker='^', label='Geometric Median Approach')

plt.title('Comparison of Performance Metrics (Delivery locations = 50)')
plt.xlabel('Scenario')
plt.ylabel('Total Distance (in K.M.)')

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

## Delivery VS Distance Calculation

```python
df = pd.read_csv("delivery_vs_distance_calculation.csv")

df.head()
```

```python
df_geometric_median = df[df["manhattan_points"]==0]
df_manhattan_median = df[df["euclidean_points"]==0]
df_global_geometric_median = df[(df['euclidean_points'] > df['manhattan_points']) & (df["manhattan_points"]!=0)]
df_global_manhattan_median = df[(df['euclidean_points'] < df['manhattan_points']) & (df["euclidean_points"]!=0)]
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(data=df_geometric_median, x='total_delivary', y='distance', marker='o', label='Geometric Median Approach')
sns.lineplot(data=df_manhattan_median, x='total_delivary', y='distance', marker='^', label='Manhattan Median Approach')
sns.lineplot(data=df_global_geometric_median, x='total_delivary', y='distance', marker='s', label='Geometric Median For Manhattan Border Approach')
sns.lineplot(data=df_global_manhattan_median, x='total_delivary', y='distance', marker='D', label='Manhattan Median For Euclidean Border Approach')


plt.title('Comparison of Performance Metrics (Delivery VS Distance Calculation)')
plt.xlabel('No. of Delivery')
plt.ylabel('Total Distance (in K.M.)')

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

## Capacity K

```python
df = pd.read_csv('./Multi_Capacity_performance.csv')

df.head()
```

```python
df_Euclidean = df[df['manhattan_points'] == 0]
df_Manhattan = df[df['euclidean_points'] == 0]
df_EM_Grid = df[(df['euclidean_points'] != 0) & (df['manhattan_points'] != 0)]
```

```python
df_Euclidean[df_Euclidean['drone_capacity'] == 4]
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(data=df_Euclidean[df_Euclidean['drone_capacity'] == 4], x='total_delivery', y='distance', marker='o', label='Multi Capacity Geometric Median Approach')
sns.lineplot(data=df_EM_Grid[df_EM_Grid['drone_capacity'] == 4], x='total_delivery', y='distance', marker='s', label='Multi Capacity EM Grid Approach')
sns.lineplot(data=df_Manhattan[df_Manhattan['drone_capacity'] == 4], x='total_delivery', y='distance', marker='^', label='Multi Capacity Manhattan Median Approach')

plt.title('No. of Delivery VS Distance (Drone Capacity = 4)')
plt.xlabel('No. of Delivery')
plt.ylabel('Total Distance (in K.M.)')

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(data=df_Euclidean[df_Euclidean['total_delivery'] == 50], x='drone_capacity', y='distance', marker='o', label='Multi Capacity Geometric Median Approach')
sns.lineplot(data=df_EM_Grid[df_EM_Grid['total_delivery'] == 50], x='drone_capacity', y='distance', marker='s', label='Multi Capacity EM Grid Approach')
sns.lineplot(data=df_Manhattan[df_Manhattan['total_delivery'] == 50], x='drone_capacity', y='distance', marker='^', label='Multi Capacity Manhattan Median Approach')

plt.title('Drone Capacity VS Distance (No. of Delivery = 50)')
plt.xlabel('No. of Delivery')
plt.ylabel('Total Distance (in K.M.)')

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
